# Attention Mechanism

Self attention allows each position in the input sequence to consider the relevancy of all the other positions in the same sequence when computing the representation of a sequence

## Simplified Attention

```python
import torch
```

```python
inputs = torch.tensor(
    [
        [0.43, 0.15, 0.89],
        [0.55, 0.87, 0.66],
        [0.57, 0.85, 0.64],
        [0.22, 0.58, 0.33],
        [0.77, 0.25, 0.10],
        [0.05, 0.80, 0.55]
    ]
)
```

### Step 1: Calculate attention scores

```python
query = inputs[1]
attn_scores_2 = torch.empty(inputs.shape[0])

for i, x_i in enumerate(inputs):
    attn_scores_2[i] = torch.dot(x_i, query) #calculate the dot product of every position with x_i

print(attn_scores_2)
```

Dot product measures how similar vectors are i.e. higher dot product == closer vectors 

### Step 2: Normalization

TThe dot product between the queries and keys produces scores that show the relationship between them. They are unnormalised and can span any range of values. This is important so as to maintain training stability in an LLM by obtaining attention weights that sum up to 1. This is done to transform the scores into a set of usable weights that represent a distribution of attention.

```python
attn_weights_2_tmp = attn_scores_2/ attn_scores_2.sum()
print('Attention weights:', attn_weights_2_tmp)
print('Sum:', attn_weights_2_tmp.sum())
```

Use the softmax in practise. It makes sure that attention weights are always positive. It is better at managing extreme values and offers more favorable gradient properties during training
